```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
calib = pd.read_csv('example_data/joycon_calibration_2023_03_06.csv')
gyro_x_calib = calib.iloc[:, 4].mean()
gyro_y_calib = calib.iloc[:, 5].mean()
gyro_z_calib = calib.iloc[:, 6].mean()

print(gyro_x_calib, gyro_y_calib, gyro_z_calib)
```

```python
data = pd.read_csv('example_data/joycon_up_2023_03_06.csv')

gx_set = data.iloc[:, 4].to_numpy() - gyro_x_calib
gy_set = data.iloc[:, 5].to_numpy() - gyro_y_calib
gz_set = data.iloc[:, 6].to_numpy() - gyro_z_calib

fig, ax = plt.subplots()

plt.plot(gx_set.cumsum())
plt.plot(gy_set.cumsum())
plt.plot(gz_set.cumsum())

plt.show()
```
